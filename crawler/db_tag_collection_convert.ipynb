# db tag 콜랙션 생성

```python
import json

with open('./db저장데이터/DB_namuwiki_docs_words.json', 'rt', encoding='utf-8') as f:
    namutag_json_data = json.load(f)

# with open('./db저장데이터/youtuber_probit.json', 'rt', encoding='utf-8') as f:
#     yourecotag_json_data = json.load(f) # DB에 존재

# with open('./db저장데이터/youtuber_sex_platform.json', 'rt', encoding='utf-8') as f:
#     sexplatform_json_data = json.load(f) # DB에 존재

with open('./db저장데이터/DB_video_data_add.json', 'rt', encoding='utf-8') as f:
    videotag_json_data = json.load(f) # 비디오 태그

with open('./db저장데이터/DB_video_data_tag.json', 'rt', encoding='utf-8') as f:
    gametag_json_data = json.load(f) # 게임 태그 

with open('./db저장데이터/comment_tag.json', 'rt', encoding='utf-8') as f:
    commenttag_json_data = json.load(f) # 게임 태그 

```

```python
from pymongo import MongoClient 
import pprint
from collections import Counter

final_json = {
    'tags' : []
}

my_client = MongoClient('localhost',27017) 
sparkdb = my_client['spark612'] 
channels_col = sparkdb['channels']

channel_col_data = channels_col.find()
for i, channel in enumerate(channel_col_data):
    # print(channel['categories'])
    # if i > 5:
    #     break

    tag_json = {
        'id' : '', #
        'title' : '', #
        'sex' : '', #
        'live_platform' : [], #
        'youreco_tags' : [], #
        'namu_tags': [], #
        'game_tags': [], 
        'video_tags' : [], #
        'comment_tags' : [], #
        'youtuber_tags' : [] #
    }
    tag_json['id'] = channel['id']

    tag_json['title'] = channel['title']

    try:
        tag_json['sex'] = channel['sex']
    except Exception as err:
        print(err)

    tag_json['live_platform'] = channel['live_platform']

    # tag_json['youreco_tags'] = channel['youreco_tags']
    for dic in channel['youreco_tags']:
        del dic['_id']
    tag_json['youreco_tags'] = channel['youreco_tags']

    try:
        tag_json['youtuber_tags'] = channel['youtuber_tags']
    except Exception as err:
        print(err)

    for j, namu_channel in enumerate(namutag_json_data['channels']):
        def key_change(x):
            x['tag'] = x.pop('text')
        if namu_channel['id'] == tag_json['id']:
            for word in namu_channel['namuwikis'][0]['cloud_words']:
                key_change(word)
            tag_json['namu_tags'] =  namu_channel['namuwikis'][0]['cloud_words']
            break

    for k, video_channel in enumerate(videotag_json_data['channels']):
        video_tags = []
        tag_json_type = []
        if video_channel['id'] == tag_json['id']:
            for video in video_channel['videos']:
                video_tags = video_tags + video['tags']
            count_tags = Counter(video_tags)
            for index, key in enumerate(list(count_tags)):
                json_data = {
                    'tag' : '',
                    'value' : 0
                }
                json_data['tag'] = key
                json_data['value'] = list(count_tags.values())[index]
                tag_json_type.append(json_data)
            tag_json['video_tags']= tag_json_type
            break

    for l, comment_channel in enumerate(commenttag_json_data['tags']):
        if comment_channel['id'] == tag_json['id']:
            tag_json['comment_tags'] = comment_channel['comment_tags']
            break

    for m, video_channel in enumerate(gametag_json_data['channels']):
        game_tags = []
        tag_json_type = []
        if video_channel['id'] == tag_json['id']:
            for video in video_channel['videos']:
                if video['game_tag'] == '':
                    continue
                game_tags.append(video['game_tag'])
            count_tags = Counter(game_tags)
            for index, key in enumerate(list(count_tags)):
                json_data = {
                    'tag' : '',
                    'value' : 0
                }
                json_data['tag'] = key
                json_data['value'] = list(count_tags.values())[index]
                tag_json_type.append(json_data)
            tag_json['game_tags']= tag_json_type
            break

            

    # pprint.pprint(tag_json)
    final_json['tags'].append(tag_json)
    print(i)

with open('DB_tags.json', 'wt',encoding='utf-8') as f:
    json.dump(final_json, f, ensure_ascii=False, indent='\t')

    
```
