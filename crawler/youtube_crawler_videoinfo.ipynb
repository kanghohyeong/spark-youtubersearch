# 영상별 동영상 정보 추출기

```python
import json
from googleapiclient.discovery import build

#api 연결
apiKey_dirt = 'APIkey.json'

with open( apiKey_dirt) as json_file:
    json_data = json.load(json_file)
    youtubeApiKey = json_data["APIkey"]

youtube = build('youtube','v3',developerKey=youtubeApiKey)

#기존 비디오 데이터 json 열기
with open('./DB_video_data_full.json', 'rt', encoding='utf-8') as f:
    video_json_data = json.load(f)


print(video_json_data['Youtuber'][0]['name'])
print(len(video_json_data['Youtuber']))
```

```python
#video 정보 가져오기
import re

stop_state = False

for i, channel in enumerate(video_json_data['Youtuber']):
    if i < 32:
        continue
    # elif i > 20:
    #     break
    if stop_state == True:
        break

    print('{} : {} start'.format(i, channel['name']))

    for j, video in enumerate(channel['Recent_videos']):
        # if j > 2 :
        #     break
        if video['like_count'] :
            print('already exist')
            continue

        try:
            api_video = youtube.videos().list(part='id, snippet,  recordingDetails, statistics,  topicDetails', id=video['video_id']).execute()
        except Exception as err:
            print('api ovver', err)
            stop_state = True
            break

        add_data = {
            'like_count' : 0,
            'dislike_count' : 0,
            'comment_count' : 0,
            'view_count' : 0,
            'video_category' : [],
            'tags' : []
        }

        try:
            video['like_count'] = api_video['items'][0]['statistics']['likeCount']
        except:
            print('{} no like count'.format(j))

        try:
            video['dislike_count'] = api_video['items'][0]['statistics']['dislikeCount']
        except:
            print('{} no dislike count'.format(j))

        try:
            video['view_count'] = api_video['items'][0]['statistics']['viewCount']
        except:
            print('{} no view count'.format(j))

        try:
            video['comment_count'] = api_video['items'][0]['statistics']['commentCount']
        except:
            print('{} no comment count'.format(j))

        try:
            for category in api_video['items'][0]['topicDetails']['topicCategories']:
                new_category = re.sub(pattern='https://en.wikipedia.org/wiki/' ,repl='', string=category)
                video['video_category'].append(new_category)
        except:
            print('{} no category '.format(j))

        try:
            for tag in api_video['items'][0]['snippet']['tags']:
                video['tags'].append(tag)
        except:
            print('no tag')


        #json 변경

        # print('{} : video clear'.format(j))
        # print(video)


with open('DB_video_data_full.json','wt',encoding='utf-8') as f:
    json.dump(video_json_data, f, ensure_ascii=False, indent='\t')

        
```
