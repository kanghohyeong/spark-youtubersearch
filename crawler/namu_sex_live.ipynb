```python
import requests
from bs4 import BeautifulSoup

def sex_platform_crawling(namu_code):
    req = requests.get('https://namu.wiki/w/'+namu_code)
    bs = BeautifulSoup(req.text, 'html.parser')

    targets = bs.find(text='분류')
    h2 = targets.parent
    ul = h2.parent.get_text()

    if len(ul) < (3+len(namu_code)+3):
        req = requests.get('https://namu.wiki/w/분류:'+namu_code)
        bs = BeautifulSoup(req.text, 'html.parser')

        targets = bs.find(text='분류')
        h2 = targets.parent
        ul = h2.parent.get_text()

    return ul

def platform_extract(namu_text):
    live_platform = ['아프리카TV', '카카오TV', '트위치']
    real_platform = []
    for platform in live_platform:
        if namu_text.find(platform) != -1:
            real_platform.append(platform)
    print(real_platform)
    return real_platform

def sex_extract(namu_text):
    list_sex = ['남성', '여성']
    real_sex = ''
    for sex in list_sex:
        if namu_text.find(sex) != -1:
            real_sex = sex
    print(real_sex)
    return real_sex

text = sex_platform_crawling('감스트')
platform_extract(text)
sex_extract(text)


```

```python
import pandas as pd
import json
import time

namu_csv = pd.read_csv('./data_with_vling/youtuber_list.csv', header=0, encoding='utf-8')

namu_csv.fillna('', inplace=True)

err_youtuber = []

result_json = {
    'channels' : []
}

for i , namu_title in enumerate(namu_csv['namu_code']):
    # if i > 2:
    #     break
    if namu_title == '':
        continue
    data_json = {
        'id' : '',
        'sex' : '',
        'live_platform' : []
    }
    data_json['id'] = namu_csv['youtube_code'][i]
    try:
        namu_text = sex_platform_crawling(namu_title)
        print(namu_text)
    except Exception as e:
        print('{} text crawling err'.format(i), e)
        err_youtuber.append(i)
        continue
    
    data_json['live_platform'] = platform_extract(namu_text)
    data_json['sex'] = sex_extract(namu_text)

    result_json['channels'].append(data_json)
    print('{} : {} ok'.format(i, namu_title))
    time.sleep(1)


    
with open('youtuber_sex_platform.json', 'wt',encoding='utf-8') as f:
    json.dump(result_json, f, ensure_ascii=False, indent='\t')
```
