```python
import json
from googleapiclient.discovery import build

#api 연결
apiKey_dirt = 'APIkey.json'

with open( apiKey_dirt) as json_file:
    json_data = json.load(json_file)
    youtubeApiKey = json_data["APIkey"]

youtube = build('youtube','v3',developerKey=youtubeApiKey)

#기존 데이터 json 열기
with open('youtuber_sex_platform.json', 'rt',encoding='utf-8') as f:
    json_data = json.load(f)




```

```python
import re

result_json = {
    'channels' : []
}

for i, channel in enumerate(json_data['channels']):
    # if i > 2:
    #     break
    db_json = {
        'id' : '',
        'sex' : '',
        'live_platform' : [],
        'profile_img' : '',
        'banner_img' : ''
    }
    db_json['id'] = re.sub(pattern='/channel/', repl='', string=channel['id']) 
    db_json['sex'] = channel['sex']
    db_json['live_platform'] = channel['live_platform']

    try:
        api_channel = youtube.channels().list(part='id, snippet, brandingSettings, contentDetails,  statistics, topicDetails', id=db_json['id']).execute()
    except Exception as err:
        print('api err', err)
        stop_state = True
        break

    try:
        db_json['profile_img'] = api_channel['items'][0]['snippet']['thumbnails']['medium']['url']
    except Exception as err:
        print('no profile_img', err)
    
    try:
        db_json['banner_img'] = api_channel['items'][0]['brandingSettings']['image']['bannerExternalUrl']
    except Exception as err:
        print('no banner img', err)

    result_json['channels'].append(db_json)
    print('{} ok'.format(i))

with open('youtuber_sex_platform_img.json','wt',encoding='utf-8') as f:
    json.dump(result_json, f, ensure_ascii=False, indent='\t')





```
