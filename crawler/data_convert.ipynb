# DB 스키마 형식으로 변환하기

```python
# 우리 db 형식으로 바꾸기

import json
import pytz
import datetime

with open('./DB_video_data_full.json', 'rt', encoding='utf-8') as f:
    video_json_data = json.load(f)

with open('./DB_data.json', 'rt',encoding='utf-8') as f:
    channel_json_data = json.load(f)


# 최종 json 

db_json = {
    'channels' : []
}



for i, channel_video in enumerate(video_json_data['Youtuber']):

    # if i > 2:
    #     break

    for j, channel_data in enumerate(channel_json_data['Youtuber']):

        if channel_video['name'] == channel_data['youtube_title']:

            db_data = {
                'id': '',
                'videos' : []
            }

            db_data['id'] = channel_data['channel_id']

        
            for k, video in enumerate(channel_video['Recent_videos']):

                video_data = {
                    'id' : '',
                    'like_num' : [{'value': 0, 'date' : str(datetime.datetime.now(tz=pytz.utc).isoformat())}],
                    'dislike_num' : [{'value': 0, 'date' :str(datetime.datetime.now(tz=pytz.utc).isoformat())}],
                    'comment_num' : [{'value': 0, 'date' : str(datetime.datetime.now(tz=pytz.utc).isoformat())}],
                    'view_num' : [{'value': 0, 'date' : str(datetime.datetime.now(tz=pytz.utc).isoformat())}],
                    'categories' : [],
                    'tags' : []
                
                }
                video_data['tags'] = video['tags']
                video_data['id'] = video['video_id']
                video_data['like_num'][0]['value'] = int(video['like_count'])
                video_data['dislike_num'][0]['value'] = int(video['dislike_count'])
                video_data['comment_num'][0]['value'] = int(video['comment_count'])
                video_data['view_num'][0]['value'] = int(video['view_count'])
                video_data['categories'] = video['video_category']

                # print(video_data)
                db_data['videos'].append(video_data)

            
   

            db_json['channels'].append(db_data)
            # print(db_json)
            break


with open('DB_video_data_add.json', 'wt',encoding='utf-8') as f:
    json.dump(db_json, f, ensure_ascii=False, indent='\t')


    
```
