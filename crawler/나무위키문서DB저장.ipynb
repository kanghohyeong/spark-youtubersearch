# 나무위키 문서 DB 저장

```python
import re

def preprocess_special(strings):
    stop_special_char = re.sub(pattern='[가-힣 a-zA-Z\d\.\n]+', repl='', string=strings)
    stop_special_char_pattern = '[' + '\\'.join(set(stop_special_char)) + ']'
    if stop_special_char_pattern == '[]':
        return strings
    new_strings = re.sub(pattern=stop_special_char_pattern, repl=' ',string=strings)
    return new_strings
def preprocess_no_eng_num(strings):
    new_strings = re.sub(pattern='[a-zA-Z]',repl='',string=strings)
    new_strings = re.sub(pattern='[0-9]', repl='', string=new_strings)
    return new_strings
def preprocess_no_num_pattern(strings):
    new_strings = re.sub(pattern='[0-9]+[ㄱ-ㅎ가-힣]*', repl=' ', string=strings)
    return new_strings

stopwords = ['라', '중', '이', '때문', '지', '이상', '등', '수', '것', '시작', '부분', '당시', '경우', '이후', '오브', '리그', '편', '위', '정도', '활동', '전', '둘',
    '한', '도중', '자체', '경기', '방송이','때','녹두','이유','문단을','대부분','번','일','리','두','초','나','문단','문','방송','적','후','단어','게임','유튜브','영상','말','역사','시청자들','본인','이전','시청자','업로드','콘텐츠','듯','현재','명','들','사람','1','개','관련','채널','모습','주','거','내','사이','자신','유튜버','컨텐츠','생각','내용','를','을','시즌','조','관','그','제목']

def count2cloud(count):
    cloud_list = []
    for word in count:
        cloud_word = {
            'text': '',
            'value': 0
        }
        cloud_word['text'] = word[0]
        cloud_word['value'] = word[1]
        cloud_list.append(cloud_word)
    return cloud_list
    
```

```python
import json
import os
import re
from konlpy.tag import Hannanum
from collections import Counter

with open('./DB_data.json', 'rt', encoding='utf-8') as f:
    channel_json_data = json.load(f)


#나무위키 문서 목록
dir_path = './data_with_vling/txt_namuwiki/'
file_list = os.listdir(dir_path)


# 최종 json 
db_json = {
    'channels' : []
}

#할당
for i, channel in enumerate(channel_json_data['Youtuber']):
    db_data = {
                'id': '',
                'namuwikis' : []
            }
    namu_data = {
        'title' : '',
        'content' : '',
        'cloud_words' : []
    }
    
    db_data['id'] = channel['channel_id']
    try:
        if(type(channel['namuwiki_title'])==type('sfds')):
            namu_data['title'] = channel['namuwiki_title']
            dir_namu_file = re.sub(pattern=':', repl='-', string=namu_data['title'])
            dir_namu_file = re.sub(pattern='/', repl='-', string=dir_namu_file)
            dir_namu_file = re.sub(pattern='\?', repl='', string=dir_namu_file)
            with open(dir_path+dir_namu_file+'.txt', 'rt', encoding='utf-8') as f:
                namu_text = f.read()
            namu_data['content'] = namu_text

    except Exception as err:
        print('no namu', err)

    # 빈도수 분석
    no_special_text = preprocess_special(namu_data['content'])
    no_num_pattern_text = preprocess_no_num_pattern(no_special_text)

    final_text = no_num_pattern_text
    h = Hannanum()
    nouns = h.nouns(final_text)
    unique_nouns = set(nouns)
    for word in unique_nouns:
        if word in stopwords:
            while word in nouns:
                nouns.remove(word)

    count = Counter(nouns)
    top_20 = count.most_common(20)
    cloud_words = count2cloud(top_20)
    namu_data['cloud_words'] = cloud_words
    




    db_data['namuwikis'].append(namu_data)
    db_json['channels'].append(db_data)
    print('{} : {} ok'.format(i, namu_data['title']))


with open('DB_namuwiki_docs_words.json', 'wt',encoding='utf-8') as f:
    json.dump(db_json, f, ensure_ascii=False, indent='\t')
```
