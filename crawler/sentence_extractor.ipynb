# 문장 추출기

## 1. 실력 문장 추출

```python
import pandas as pd

words = ['브론즈','실버','골드','플레티넘','다이아몬드','첼린저','천상계','실력' ,'다이아','챌린저','프로','우승','마스터','그랜드마스터','최상위','고수','프로게이머','피지컬','장인','고티어','그랜드 마스터','티어','난이도','양학','압살','압도','컨트롤','숙련도','승률','랭킹','답답']
stopwords = ['버튼','구독자','프로젝트','프로필','프로그램','다이아 페스티벌',' 다이아TV','다이아tv',' 다이아 TV',' 다이아 tv','프로젝터','아이 패드','아이폰','마스터링','마스터리','리마스터']

data = pd.read_csv('channels.csv')
names = data['name_namuwiki']

name_list = []
data_list = []

for name in names:
    print(name)
    data = pd.read_csv('namuwiki_csv/'+name+'.csv', encoding = 'utf-8')
    data = data["sentences"]
    for sent in data:
        for word in words:
            if sent.find(word) != -1:
                data_list.append(sent)
                name_list.append(name)

                for stopword in stopwords:
                    if sent.find(stopword) != -1:
                        data_list.pop()
                        name_list.pop()
                        break
                break
df = pd.DataFrame({"name":name_list , "sentence":data_list})
print(df)
df.to_csv("/extracted_sentence_csv/test-sentences-ability.csv",index  = False, encoding = 'utf-8')
```
