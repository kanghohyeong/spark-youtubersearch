# 데이터 특성 분석

```python
import json

with open('./DB_video_data_full.json', 'rt', encoding='utf-8') as f:
    video_json_data = json.load(f)

with open('./DB_data.json', 'rt',encoding='utf-8') as f:
    channel_json_data = json.load(f)


```

```python
import re

val_data = {
    'channel_desc' : 0,
    'channel_keywords' : 0,
    'video_tags' : 0,
    'video_desc' : 0,
    'total_docs' : 0
}

for i, channel_video in enumerate(video_json_data['Youtuber']):

    if i > 178:
        break

    info_data = {
        'channel_desc' : '',
        'channel_keywords' : '',
        'video_tags' : [],
        'video_titles' : '',
        'video_descs' : ''
    }



    for j, channel_data in enumerate(channel_json_data['Youtuber']):
        if channel_video['name'] == channel_data['youtube_title']:
            info_data['channel_desc'] = channel_data['channel_desc']
            info_data['channel_keywords'] = channel_data['keywords']
            
            for k, video in enumerate(channel_video['Recent_videos']):
                info_data['video_tags'] = info_data['video_tags'] + video['tags']
                info_data['video_titles'] = info_data['video_titles'] + ' ' + video['video_title']
                info_data['video_descs'] = info_data['video_descs'] + ' ' + video['video_desc']
            
            info_data['video_tags'] = list(set(info_data['video_tags']))
            # print(info_data)

            data2text = info_data['channel_desc'] + info_data['channel_keywords'] + info_data['video_titles'] + info_data['video_descs'] + ' '.join(info_data['video_tags'])
            # print(data2text)

            if info_data['channel_desc'] != '':
                val_data['channel_desc'] += 1
            if info_data['channel_keywords'] != '' :
                val_data['channel_keywords'] += 1
            if info_data['video_tags'] != [] :
                val_data['video_tags'] += 1
            if info_data['video_descs'] != '':
                val_data['video_desc'] += 1
            val_data['total_docs'] += 1
            print(val_data)

            title = re.sub(pattern='[?|/]', repl='_', string=channel_video['name'])
            print(title)

            with open('./youtube_text_data/'+title + '.txt', 'wt', encoding='utf-8') as f:
                f.write(data2text)

            break


    
```
