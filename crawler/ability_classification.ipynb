# 유튜버 실력 있는지 없는지 분류기

```python
# 새로운 문장 예측함수
import pickle

def using_model(string):
    import numpy as np
    with open('pipe_full-tfidf-2.dat', 'rb') as fp:
        pipeline = pickle.load(fp)

    text = [string]
    

    # 예측 스코어
    r1 = np.max(pipeline.predict_proba(text) * 100)
    
    # 예측 결과
    r2 = pipeline.predict(text)[0]
    
    # print("[" + string + "]" + "문장은")
    if r2 == 1 :
        print('실력 좋다에 가깝다')
    else :
        print('실력 나쁘다에 가깝다')

    print('스코어 : %.3f '% r1)
    print("*************************\n")


```

```python
# 실력에 관한 문장만 추출해서 예측

import pandas as pd

sentence_df = pd.read_csv('ability_sentences.csv', encoding='utf-8')

for sentence in sentence_df['sentence']:
    print("[" + sentence + "]" + "문장은")
    using_model(sentence)
```

```python
# 전처리 함수
def cleaning(text):
    import re
    text = re.sub(pattern="[^ㄱ-ㅎ 가-힣 a-z A-Z]", repl=" ", string=text)
    # print(text)

    import nltk
    from konlpy.tag import Okt
    t = Okt()

    stop_words = ['을', '를','의','이','에','로','가','들','으로','은','도','에서','했다','것','인','는','와','있다','때','과','그','하고','한다','년','수','이다','하는','한','할','등','월','일','까지','에게','다']

    morph_text = t.morphs(text)
    clean_text = [each_word for each_word in morph_text
                if each_word not in stop_words]
    final_text = " ".join(clean_text)

    # print(final_text)
    return final_text

```
